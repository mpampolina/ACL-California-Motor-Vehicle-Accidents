# Filter-US-Accident-CSV
The following code is used to trim a master accident dataset, found here https://bit.ly/3dO7Kiy, into a more manageable dataset targetted at accidents occuring in the US state of California.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import json
```

```python
path = os.path.join('data', 'US_Accidents_Feb16_Dec19.csv')
df_lrg = pd.read_csv(path)
```

```python
pd.set_option('display.max_columns', None)
```

```python
df_lrg
```

```python
filt = df_lrg['State'] == 'CA'
df = df_lrg.loc[filt]
df
```
